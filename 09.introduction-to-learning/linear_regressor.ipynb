```python
import numpy as np
```

```python
rng = np.random.default_rng()
```

```python
slope = rng.uniform(-5.0, 5.0)
y_intercept = rng.uniform(-10.0, 10.0)
```

```python
X = rng.uniform(-10, 10, size=100)
noise = rng.normal(loc=0.0, scale=2.0, size=100)
```

```python
y = slope * X + y_intercept + noise
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter(x=X, y=y, mode='markers')])
fig.add_hline(y=0, line_color='black', line_width=1)
fig.add_vline(x=0, line_color='black', line_width=1)
fig.show()
```

```python
X = X.reshape(-1, 1)
X = np.insert(X, 0, 1, axis = 1)
y = y.reshape(-1, 1)
```

```python
pred_y_intercept, pred_slope = np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
X_pred = X[:, 1]
y_pred = pred_slope * X_pred + pred_y_intercept
```

```python
fig.add_trace(go.Scatter(x=X_pred, y=y_pred, mode='lines'))
fig.show()
```
